考虑问题的表述，很显然，决策就是选择综合公寓大楼的最佳规模。PDC有如下三个决策方案

在决策分析中，一个偶然事件的可能结果被称为**自然状态**。对于PDC公司的问题，与公寓需求相关的偶然有两种自然状态：  
$s_1$——对公寓的强需求

# 21.2 概率决策
接下来我们可以集中精力确定自然状态的概率。古典方法、相对频数方法或第4章所讨论的分配概率的主观方法都可以用来确定这些概率  
在决定了合适的概率之后，我们将介绍如何运用**期望值法**来确定问题最佳的或建议的决策方案  
## 21.2.1 期望值法
我们从定义一个决策方案的期望值开始我们的讨论。设  
N——自然状态的个数  
P($s_j$)——自然状态$s_j$的概率 5  
由于在N种自然状态种有且仅有一种自然状态能够发生，因此概率必须满足两个条件(21-1)(21-2)  
$$P(s_j)\ge0，对所有的自然状态$$
$$\sum_{j=1}^N{P(s_j)}=P(s_1)+P(s_2)+\dots+P(s_N)=1$$
决策方案$d_i$的**期望值**如下  
<hr />

**期望值**（21-3）  
$$EV(d_i)=\sum_{j=1}^N{P(s_j)V_{ij}}$$
5  
式中，$V_{ij}$代表决策方案$d_i$和自然状态$s_j$的支付值  
<hr />

简而言之，一个决策方案的期望值是该决策方案支付的加权和  
假定这种乐观态度导致强需求的初始主观概率的评估值为0.8，相应的弱需球（$s_2$）的初始主观概率的评估值为0.2  
利用表21-1的支付值和式（21-3），我们计算出三个决策方案的期望值如下：  

```python
'''
5
'''
evd1=0.8*8+0.2*7
evd2=0.8*14+0.2*5
evd3=0.8*20+0.2*(-9)
print('{:.1f},{:.1f},{:.1f}'.format(evd1,evd2,evd3))
```

因此，利用期望值法，我们发现大型综合公寓大楼的期望值为1420万美元，是推荐的方案  
用于确定具有最佳期望值的决策方案所需要的计算可以方便的通过决策树来完成  
图21-2表示的是具有自然状态分枝概率的PDC问题的决策树。正如图21-3所示，通过这样计算，我们得到节点2、3、4的期望值  
![21-2](../syn_pic/statistics_for_business_economics/21-2.png)  
![21-3 5](../syn_pic/statistics_for_business_economics/21-3.png)  
由于决策者控制了从决策节点1延伸出来的分枝，并且我们正在试图使利润最大化，所以节点1的最佳决策方案就是$d_3$  
首先，绘出由决策节点、机会节点和描述问题连续性的分枝构成的决策树  
如果你用期望值法，下一步是决定每种自然状态的概率及计算每个机会节点的期望值  
然后选择导致具有最佳期望值的机会节点的决策分支。与这一分枝相联系的决策方案是推荐的决策  
## 21.2.2 完全信息的期望值
假定PDC有机会进行一次市场调研，该调研有助于评估买家对公寓大楼项目的兴趣，并能为管理层提供改善自然状态概率评估的信息 5  
一个决策策略不过是一个决策规则，该决策规则在新的信息可以使用后，提出可供选择的决策方案  
为了帮助PDC确定决策策略，我们复制PDC支付表，如表21-1所示  
![tb21-1](../syn_pic/statistics_for_business_economics/tb21-1.png)  
因此，如果完全信息是可以使用的，我们就可以描述PDC的最优决策策略如下：  
如果是自然状态$s_1$，则选择$d_3$，并且得到2000万美元支付  
如果是自然状态$s_2$，则选择$d_1$，并且得到700万美元支付 5  
要计算完全信息的期望值，让我们回到自然状态的初始概率：此时$P(s_1)=0.8$、$P(s_2)=0.2$  
因此，运用式（21-3），以完全信息为依据的决策策略的期望值就是  

```python
0.8*20+0.2*7
```

我们把期望值1740万美元称为完全信息条件下的期望值  
本节前面的决策建议和期望值计算没有完全信息的帮助，因此我们把1420万美元称为不完全信息条件下的期望值 5  
如果有关自然状态的信息是完全可使用的，那么320万美元表示能得到的额外期望值  
一般而言，一个市场调研并不会提供“完全”的信息；然而，如果是一个好的市场调研，收集的信息的价值可能占320万美元相当大的比例  
一般而言，**完全信息的期望值**的计算如下：  
<hr />

**完全信息期望值**（21-4）  
$$EVPI=|EVwPI-EVwoPI|$$  
5  
式中，EVPI代表完全信息期望值;EVwPI代表自然状态的完全信息条件下的期望值;EVwoPI代表自然状态的不完全信息条件下的期望值  
<hr />

注意式（21-4）种绝对值的作用 5  
# 21.3 样本信息的决策分析
通常情况下，决策者对自然状态已经有初步概率或**先验概率**的估计，这个概率估计值是当时可以使用的最好概率值  
在大多数情形下，通过实验涉及提供的自然状态的样本信息，我们能得到自然状态的额外信息。这些修正后的概率被称为**后验概率**

1. 有利报告。相当数量的调查者表示对购买PDC公寓有兴趣

# 关键术语
**偶然事件** 影响与一个决策相关的结果或支付的不确定的未来事件  
**结果** 当选定决策方案且偶然事件发生时所获得的结果。结果的度量通常称为支付  
**自然状态** 影响决策方案支付的偶然事件的可能结果  
**支付** 决策结果的度量  
**支付表** 决策问题支付的列表表示 5  
**决策树** 显示决策制定过程连续性的决策问题的图形表示  
**节点** 一个影响图示或决策树的交叉点或结合点  
**决策节点** 表示要制定决策的节点  
**机会节点** 表示一个不确定事件将要发生的节点  
**分枝** 表示决策节点的方案与机会节点的结果的一段线 5  
**期望值法** 以每一决策方案的期望值为依据的选择决策方案的方法  
**期望值** 对于一个机会节点，它是支付的加权平均值。权重是自然状态的概率  
**完全信息的期望值** 能准确地告知决策者哪种自然状态将要发生信息地期望值  
**先验概率** 获得样本信息前自然状态的概率  
**样本信息** 通过能够更新或修正自然状态概率的研究或实验而获得的新信息 5  
**后验（修正）概率** 以样本信息为依据，修正了先验概率后的自然状态概率  
**决策策略** 一个策略，包括为决策问题提供最优解的一系列的决策和偶然结果  
**样本信息的期望值** 以样本信息为依据的最优策略的期望值与没有任何样本信息的“最佳”期望值之间的差  
**贝叶斯定理** 一个能利用样本信息来修正先验概率的定理  
**条件概率** 一个事件在已知相关事件结果的条件下的概率 5  
**联合概率** 样本信息与一个具体的自然状态同时发生的概率 5
